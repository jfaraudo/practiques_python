# Exemples Algebra

```python
import numpy as np
```

* Exemples de vectors (array)

```python
v=np.array([1,0,0])
u=np.array([0,-2,0])
```

```python
print(u)
```

Operacions amb vectors (suma, resta, producte escalar i vectorial)

```python
w1=u+v
print(w1)
w2=u-v
print(w2)
```

```python
a=np.dot(u,v)
print(a)
```

```python
b=np.cross(v,u)
print(b)
```

* Exemples amb Matrius

```python
A=np.array([[2,4],[1,3]])
print(A)
```

```python
e=A[1,0]
print(e)
```

```python
v=A[0,:]
print(v)
```

```python
u=A[:,1]
print(u)
```

Valors i vectors propis

```python
A=np.array([[2,2],[1,3]])
```

```python
Valors,Vectors = np.linalg.eig(A)
print("Valors propis", Valors)
print("Vectors propis",Vectors)
```

```python
u=Vectors[:,0]
print(u)
```

```python
v=Vectors[:,1]
print(v)
```

comprovem que efectivament son vectors propis 

```python
z=np.dot(A,u)
print(z)
```

```python
z2=np.dot(A,v)
print(z2)
```
